# Q7: Create a visualization with a software/tool of your choice for age and age by gender distribution of patients in the dataset. Paste the visualizations in your answer. Also, briefly describe the distributions.

Assumption 5: Age is calculated as the difference between ADMISSION_DATE and BIRTHDATE. If a patient visited hospital multiple times, take age at the last visit (maximum available age).

```python
# drop time from BIRTHDATE column (cannot convert BIRTHDATE to datetime format from its original form)
df_merge['BIRTHDATE_DATE'] = df_merge.BIRTHDATE.str.split(':').str[0]

# convert values in BIRTHDATE_DATE column to datetime
df_merge.BIRTHDATE_DATE = pd.to_datetime(df_merge.BIRTHDATE_DATE)
```

```python
# review BIRTHDATE_DATE column for Y2K issue
df_merge.BIRTHDATE_DATE.dt.year.value_counts()
```

Since BIRTHDATE column only had 2 digits for a year, some years were converted into 20XX and others into 19XX. Looks like the cutoff point is year 71.

```python
# function to convert birthday years into 19XX format
def fix_date(x):
    if x.year > 2015:  
        year = x.year - 100  # subtract 100 years if BIRTHDATE_DATE year is greated than 2015 
    else:
        year = x.year        # keep the same year if BIRTHDATE_DATE year is less than or equal to 2015 

    return datetime.date(year,x.month,x.day)
```

```python
# apply function to convert birthday years into 19XX format
df_merge.BIRTHDATE_DATE = df_merge.BIRTHDATE_DATE.apply(fix_date)

# convert values in BIRTHDATE_DATE column to datetime
df_merge.BIRTHDATE_DATE = pd.to_datetime(df_merge.BIRTHDATE_DATE)
```

```python
# calculate age as a difference between ADMISSION_DATE and BIRTHDATE_DATE in a new column
df_merge['AGE'] = (df_merge.ADMISSION_DATE - df_merge.BIRTHDATE_DATE)// np.timedelta64(1, 'Y')

# examine the first 10 rows
df_merge.head(10)
```

```python
# get Patient IDs and age at the last visit (max available age)
df_merge.groupby(['PATIENT_ID'])['AGE'].max()
```

```python
# set the figure size
fig, ax = plt.subplots(figsize = (10,5))

# set up the histogram
ax.hist(df_merge.groupby(['PATIENT_ID'])['AGE'].max(), bins = 14)

# set up the labels
ax.set_title('Age Distribution of Patients in the Dataset')
ax.set_xlabel('Age')
ax.set_ylabel('Number of Patients')
ax.set_ylim(bottom=0, top=20)
ax.set_xlim(left=0, right=120);
```

The majority of patients are between the ages of 45 and 60. The next largest group is patients aged 60+, followed by patients aged between 20 and 45. There're no patients under 20 in the dataset.

# Q9: A) Investigate inpatient visits (patients who stayed for more than a day at the hospital) for each patient. Output a list of the top 10 patient_ids who spent maximum days in the hospital in the past year (2015). Paste the output table in your answer.

```python
# calulate the difference between DISCHARGE_DATE_TIME and January 1, 2015
df_merge['HOSP_DAYS_2015'] = (pd.to_datetime(df_merge.DISCHARGE_DATE_TIME.str.split(':').str[0]) -
                         pd.to_datetime('2014/12/31')) / np.timedelta64(1, 'D')
```

```python
# function to convert negative HOSP_DAYS_2015 into 0
def fix_hosp_days_2015(x):
    if x < 0:  
        x = 0  # assign 0 if HOSP_DAYS_2015 < 0 
    else:
        x = x  # keep the same year if HOSP_DAYS_2015 >= 0 

    return x
```

```python
# apply function to convert negative HOSP_DAYS_2015 into 0
df_merge.HOSP_DAYS_2015 = df_merge.HOSP_DAYS_2015.apply(fix_hosp_days_2015)
```

```python
# to account for patints who were admitted in 2015, replace HOSP_DAYS_2015 with HOSP_DAYS
# in other words, earlier we calculated number of days between 12/31/2014 and discharge date. Now we replace it with 
# the number of days between discharge data and admission date (if the admission date is in 2015)
df_merge.HOSP_DAYS_2015 = np.where(df_merge.HOSP_DAYS_2015 > df_merge.HOSP_DAYS, 
                                   df_merge.HOSP_DAYS, df_merge.HOSP_DAYS_2015)

# examine the first 10 rows
df_merge.head(10)
```

```python
# sort the largest 10 sums of days at the hospital in 2015 by patient
df_merge.groupby(['PATIENT_ID'])['HOSP_DAYS_2015'].sum().sort_values()[-10:]
```
